```python
import matplotlib.pyplot as plt
import numpy as np
import pickle as rick
%matplotlib widget
from ipywidgets import *
from matplotlib.widgets import Slider, Button
from matplotlib.animation import FuncAnimation

def t_max(s, m):
    return 0.5 * (m_h ** 2 - s + 2 * m ** 2 + (m_h ** 2 - s) * (1 - (4 * m ** 2) / s) ** 0.5) 

def t_min(s, m):
    return 0.5 * (m_h ** 2 - s + 2 * m ** 2 - (m_h ** 2 - s) * (1 - (4 * m ** 2) / s) ** 0.5) 

alpha = 1 / 132.184
m_W = 80.379
m_Z = 91.1876
m_e = 0.511E-3
m_m = 105.66E-3
m_h = 125.1

m = m_m
E_phot_min = 5 #GeV
s_min_fix = 4 * m ** 2
s_max_fix = m_h ** 2
t_min_fix = m ** 2
t_max_fix = (m_h - m) ** 2

e = (4 * np.pi * alpha) ** 0.5
cos = m_W / m_Z
sin = np.sqrt(1 - cos ** 2)
v = 2 * m_W * sin / e
Gam_Z = 2.4952
C_eB = 10 ** -10
C_eW = C_eB
C_eH = 10 ** -10
C_HB = 2 * 10 ** -10
C_HW = 10 ** -10
#Pick C_HD and C_HWB s.t. we can use the SM vev
C_HD = 10 ** -10
C_HWB = C_HD
C_AA = cos ** 2 * C_HB + sin ** 2 * C_HW - sin * cos * C_HWB
C_AZ = 2 * sin * cos * (C_HW - C_HB) - (cos ** 2 - sin ** 2) * C_HWB
v_alt = v / (1 + v ** 2 * (cos / sin) * (C_HWB + (cos / (4 * sin)) * C_HD))
m = m_m
bins = 100

def diff_decay2(s, t, ml, C_eB, C_eW, C_AA, C_AZ, C_eH, SW, CW):
    N = 1 / (32 * (2 * np.pi) ** 3  * m_h ** 3)
    u = m_h ** 2 + 2 * ml ** 2 - s - t
    struc_4 = (4 * e ** 2 * (m_h ** 4 + s ** 2)) / (t * u) #1/v^2
    
    f6_AA = (-64 * C_AA * ml ** 2 * m_W ** 2) / (v ** 2 * s * (t - ml ** 2) * (u - ml ** 2)) #1/v^2
    struc_6_AA = f6_AA * (-(t - ml ** 2) ** 3 - (u - ml ** 2) ** 3 + (t - ml ** 2) ** 2 * (u - ml ** 2) + (t - ml ** 2) * (u - ml ** 2) * (u - ml ** 2 - 4 * m_h ** 2) + 4 * ml ** 2 * (u + t - 2 * ml ** 2) ** 2)
    f8_AA = (64 * C_AA ** 2 * m_W ** 2 * SW ** 2) / s
    struc_8_AA = f8_AA * ((u - ml ** 2) ** 2 + (t - ml ** 2) ** 2)
    
    f6_AZ = ((-8 * C_AZ * ml ** 2 * m_W ** 2 * SW ** 2 * (CW ** 2 - 3 * SW ** 2)) / (SW * CW * v ** 2 * (u - ml ** 2) * (t - ml ** 2))) * ((m_Z ** 2 - Gam_Z ** 2 - s) / ((m_Z ** 2 - Gam_Z ** 2 - s) ** 2 + 4 * m_Z ** 2 * Gam_Z ** 2)) #1/v^2
    struc_6_AZ = f6_AZ * (-(t - ml ** 2) ** 3 - (u - ml ** 2) ** 3 + (t - ml ** 2) ** 2 * (u - ml ** 2) + (t - ml ** 2) * (u - ml ** 2) * (u - ml ** 2 - 4 * m_h ** 2) + 4 * ml ** 2 * (u + t - 2 * ml ** 2) ** 2)
    f8_AZ = (2 * C_AZ ** 2 * m_W ** 2) / (CW ** 2 * (Gam_Z ** 2 * m_Z ** 2 + (m_Z ** 2 - s) ** 2))
    struc_8_AZ = f8_AZ * (CW ** 4 - 2 * CW ** 2 * SW ** 2 + 5 * SW ** 4) * s * ((u - ml ** 2) ** 2 + (t - ml ** 2) ** 2)
    
    f8_AA_AZ = ((16 * C_AA * C_AZ * m_W ** 2 * SW * (1 - 4 * SW ** 2)) / (CW)) * ((m_Z ** 2 - Gam_Z ** 2 - s) / ((m_Z ** 2 - Gam_Z ** 2 - s) ** 2 + 4 * m_Z ** 2 * Gam_Z ** 2))
    struc_8_AA_AZ = f8_AA_AZ * ((u - ml ** 2) ** 2 + (t - ml ** 2) ** 2)
    
    f6_eH = (-16 * 2 ** 0.5 * ml * m_W ** 2 * SW ** 2 * C_eH) / (v * (u - ml ** 2) * (t - ml ** 2)) #1/v
    struc_6_eH = f6_eH * (2 * m_h ** 4 - 2 * m_h ** 2 * (u + t - 2 * ml ** 2) + (u + t - 2 * ml ** 2) ** 2)
    f8_eH = (8 * m_W ** 2 * SW ** 2 * v ** 2 * C_eH ** 2) / ((u - ml ** 2) * (t - ml ** 2)) #v^2
    struc_8_eH = f8_eH * (2 * m_h ** 4 - 2 * m_h ** 2 * (u + t - 2 * ml ** 2) + (u + t - 2 * ml ** 2) ** 2)
    
    f8_eBeW_o = 1 
    struc_8_eBeW_o = f8_eBeW_o * (- 16 * SW * m_h ** 2 * m_W * v * C_eH * (SW * C_eW - CW * C_eB)) #v
    
    struc_8_eBeW = 8 * (u - ml ** 2) * (t - ml ** 2) * (SW * C_eW - CW * C_eB) ** 2 
    
    f6_eBeW = (m_h ** 2 * m_W * SW * ml) / (v ** 2) #1/v^2
    struc_6_eBeW = f6_eBeW * (16 * 2 ** 0.5 + (8 * 2 ** 0.5 * ml * CW) / (u)) * (SW * C_eW - CW * C_eB)
    
    return np.concatenate((N * (ml / v) ** 2 * np.array([struc_4]), N * np.array([struc_6_AA, struc_6_AZ, struc_6_eH, struc_6_eBeW]), N * np.array([struc_8_AA, struc_8_AZ, struc_8_eH, struc_8_AA_AZ, struc_8_eBeW, struc_8_eBeW_o])), axis=0)

def scale(l, i_HB, i_HW, i_HWB, i_eH, i_eB, i_eW):
    CW_alt = cos * (1 + (v ** 2 / 4) * C_HD + ((sin * v ** 2) / (2 * sin)) * (m / (v * 10 ** 6)) * i_HB)
    SW_alt = sin * (1 - ((cos ** 2 * v ** 2) / (4 * sin ** 2)) * C_HD - ((cos * v ** 2) / (2 * sin)) * (m / (v * 10 ** 6)) * i_HB)
    v_alt = v / (1 + v ** 2 * (cos / sin) * ((m / (v * 10 ** 6)) * i_HWB + (cos / (4 * sin)) * C_HD))
    init = (m / (v * 10 ** 6)) * np.array([cos ** 2 * i_HB + sin ** 2 * i_HW - sin * cos * i_HWB, 2 * sin * cos * (i_HB - i_HW) - (cos ** 2 - sin ** 2) * i_HWB, i_eH, i_eB, i_eW])
    I = []
    
    SM4 = l[0] * (v / v_alt) ** 2
    AA6 = (init[0] / C_AA) * l[1] * (v / v_alt) ** 2
    AZ6 = (init[1] / C_AZ) * l[2] * (v / v_alt) ** 2 * (((SW_alt ** 2 * (CW_alt ** 2 - 3 * SW_alt ** 2)) / (SW_alt * CW_alt)) / ((sin ** 2 * (cos ** 2 - 3 * sin ** 2)) / (sin * cos)))
    eH6 = (init[2] / C_eH) * l[3] * (v / v_alt) * (SW_alt / sin) ** 2
    #eBeW6 = ((sin * init[4] - cos * init[3]) / (sin * C_eW - cos * C_eB)) * l[4] * (v / v_alt) ** 2
    eBeW6_u = (((l[4] / ((sin * C_eW - cos * C_eB) * ((m_h ** 2 * m_W * sin * m) / (v ** 2)))) - 16 * 2 ** 0.5) / (8 * 2 ** 0.5 * m * cos)) ** -1
    eBeW6 = ((m_h ** 2 * m_W * SW_alt * m) / (v_alt ** 2)) * (16 * 2 ** 0.5 + (8 * 2 ** 0.5 * m * CW_alt) / eBeW6_u) * (SW_alt * C_eW - CW_alt * C_eB)
    AA8 = (init[0] ** 2 / C_AA ** 2) * l[5] * (SW_alt / sin) ** 2
    AZ8 = (init[1] ** 2 / C_AZ ** 2) * l[6] * (((CW_alt ** 4 - 2 * CW_alt ** 2 * SW_alt ** 2 + 5 * SW_alt ** 4) / CW_alt ** 2) / ((cos ** 4 - 2 * cos ** 2 * sin ** 2 + 5 * sin ** 4) / cos ** 2))
    eH8 = (init[2] ** 2 / C_eH ** 2) * l[7] * (v_alt / v) ** 2 * (SW_alt / sin) ** 2
    AAAZ8 = ((init[0] * init[1]) / (C_AA * C_AZ)) * l[8] * (((SW_alt * (1 - 4 * SW_alt ** 2)) / CW_alt) / ((sin * (1 - 4 * sin ** 2)) / cos))
    eBeWo8 = ((init[2] * SW_alt * (SW_alt * init[4] - CW_alt * init[3])) / (C_eH * sin * (sin * C_eW - cos * C_eB))) * l[10] * (v_alt / v)
    eBeW8 = ((SW_alt * init[4] - CW_alt * init[3]) ** 2 / (sin * C_eW - cos * C_eB) ** 2) * l[9]
    
    I.append(SM4)
    I.append(AA6)
    I.append(AZ6)
    I.append(eH6)
    I.append(eBeW6)
    I.append(AA8)
    I.append(AZ8)
    I.append(eH8)
    I.append(AAAZ8)
    I.append(eBeWo8)
    I.append(eBeW8)
    return I

def scale_anim_alt(l, C_HBs, C_HWs, C_HWBs, C_eHs, C_eBs, C_eWs):
    I = []
    k = (m / v) * 10 ** -6
    for i in range(0, len(C_HBs)):
        #print(scale(l, C_HBs[i] / k, C_HWs[i] / k, C_HWBs[i] / k, C_eHs[i] / k, C_eBs[i] / k, C_eWs[i] / k)[1])
        I.append(scale(l, C_HBs[i] / k, C_HWs[i] / k, C_HWBs[i] / k, C_eHs[i] / k, C_eBs[i] / k, C_eWs[i] / k))
    return np.array(I)

def scale_anim(l, C_HBs, C_HWs, C_HWBs, C_eHs, C_eBs, C_eWs):
    I = []
    for i in range(0, len(C_HBs)):
        CW_alt = cos * (1 + (v ** 2 / 4) * C_HD + ((sin * v ** 2) / (2 * sin)) * C_HBs[i])
        SW_alt = sin * (1 - ((cos ** 2 * v ** 2) / (4 * sin ** 2)) * C_HD - ((cos * v ** 2) / (2 * sin)) * C_HBs[i])
        v_alt = v / (1 + v ** 2 * (cos / sin) * (C_HWBs[i] + (cos / (4 * sin)) * C_HD))
        
        C_AA_i = cos ** 2 * C_HBs[i] + sin ** 2 * C_HWs[i] - sin * cos * C_HWBs[i]
        C_AZ_i = 2 * sin * cos * (C_HWs[i] - C_HBs[i]) - (cos ** 2 - sin ** 2) * C_HWBs[i]
        
        #SM = l[0] * (v / (v / (1 + v ** 2 * (cos / sin) * ((C_HWBs[i]) + (cos / (4 * sin)) * C_HD)))) ** 2
        SM = l[0] * (v / v_alt) ** 2
        AA6 = (C_AA_i / C_AA) * l[1] * (v / v_alt) ** 2
        AZ6 = (C_AZ_i / C_AZ) * l[2] * (v / v_alt) ** 2 * (((SW_alt ** 2 * (CW_alt ** 2 - 3 * SW_alt ** 2)) / (SW_alt * CW_alt)) / ((sin ** 2 * (cos ** 2 - 3 * sin ** 2)) / (sin * cos)))
        eH6 = (C_eHs[i] / C_eH) * l[3] * (v / v_alt) * (SW_alt / sin) ** 2
        #eBeW6 = ((sin * C_eWs[i] - cos * C_eBs[i]) / (sin * C_eW - cos * C_eB)) * l[4] * (v / v_alt) ** 2
        eBeW6_u = (((l[4] / ((sin * C_eWs[i] - cos * C_eBs[i]) * ((m_h ** 2 * m_W * sin * m) / (v ** 2)))) - 16 * 2 ** 0.5) / (8 * 2 ** 0.5 * m * cos)) ** -1
        eBeW6 = ((m_h ** 2 * m_W * SW_alt * m) / (v_alt ** 2)) * (16 * 2 ** 0.5 + (8 * 2 ** 0.5 * m * CW_alt) / eBeW6_u) * (SW_alt * C_eWs[i] - CW_alt * C_eBs[i])
        AA8 = (C_AA_i ** 2 / C_AA ** 2) * l[5] * (SW_alt / sin) ** 2
        AZ8 = (C_AZ_i ** 2 / C_AZ ** 2) * l[6] * (((CW_alt ** 4 - 2 * CW_alt ** 2 * SW_alt ** 2 + 5 * SW_alt ** 4) / CW_alt ** 2) / ((cos ** 4 - 2 * cos ** 2 * sin ** 2 + 5 * sin ** 4) / cos ** 2))
        eH8 = (C_eHs[i] ** 2 / C_eH ** 2) * l[7] * (v / v_alt) ** -2 * (SW_alt / sin) ** 2
        AAAZ8 = ((C_AA_i * C_AZ_i) / (C_AA * C_AZ)) * l[8] * (((SW_alt * (1 - 4 * SW_alt ** 2)) / CW_alt) / ((sin * (1 - 4 * sin ** 2)) / cos))
        eBeWo8 = ((C_eHs[i] * (SW_alt * C_eWs[i] - CW_alt * C_eBs[i])) / (C_eH * (sin * C_eW - cos * C_eB))) * l[10] * (v / v_alt) ** -1
        eBeW8 = ((SW_alt * C_eWs[i] - CW_alt * C_eBs[i]) ** 2 / (sin * C_eW - cos * C_eB) ** 2) * l[9]
        I.append(np.concatenate((SM, AA6, AZ6, eH6, eBeW6, AA8, AZ8, eH8, AAAZ8, eBeW8, eBeWo8), axis = 0))
    return np.array(I)
```
